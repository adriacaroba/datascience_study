## Fazendo as primeiras visualizações

Nós trabalharemos com dados de uma estação de meteorologia, que registra informações como temperatura, pressão, umidade e assim por diante. Foi pedido um relatório que analisasse a temperatura ao longo do tempo a partir desses dados.

```python
import pandas as pd
```

```python
df = pd.read_csv('dados/monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info()
```

- **Convertendo data (string) para datatime**

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#gráfico de linhas
plt.plot(df['data'], df['temperatura'])
```

```python
#aumentando o tamanho do gráfico
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
```

```python
#adicionando um título
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo')
```

- **Funcionalidades básicas do matplotlib**

```python
#criando um objeto de figura
fig = plt.figure(figsize=(15,8))
axes = fig.add_axes([0,0,1,1])
axes.plot(df['data'], df['temperatura'])
axes.set_title('Temperatura no momento')
```

```python
#adicionando legenda nos eixos
fig = plt.figure(figsize=(15,8))
axes = fig.add_axes([0,0,1,1])
axes.plot(df['data'], df['temperatura'])
axes.set_title('Temperatura no momento')
axes.set_ylabel('Temperatura')
axes.set_xlabel('Data')
```

```python
#aumentando o tamanho da fonte
fig = plt.figure(figsize=(15,8))
axes = fig.add_axes([0,0,1,1])
axes.plot(df['data'], df['temperatura'])
axes.set_title('Temperatura no momento', fontsize=25)
axes.set_ylabel('Temperatura', fontsize=20)
axes.set_xlabel('Data', fontsize=20)
```

```python
#adicionando legenda do dados
fig = plt.figure(figsize=(15,8))
axes = fig.add_axes([0,0,1,1])
axes.plot(df['data'], df['temperatura'])
axes.set_title('Temperatura no momento', fontsize=25)
axes.set_ylabel('Temperatura', fontsize=20)
axes.set_xlabel('Data', fontsize=20)
axes.legend(['temperatura'], loc='upper right', fontsize = 15)
```

```python
#adicionando legenda do dados
fig = plt.figure(figsize=(15,8))
axes = fig.add_axes([0,0,1,1])
axes.plot(df['data'], df['temperatura'], color = 'green')
axes.set_title('Temperatura no momento', fontsize=25)
axes.set_ylabel('Temperatura', fontsize=20)
axes.set_xlabel('Data', fontsize=20)
axes.legend(['temperatura'], loc='upper right', fontsize = 15)
```

# Aula 3

## Anotações e diversos gráficos

- **Linhas de restrição e anotações**

```python
#adicionando linhas de restrição
fig = plt.figure(figsize = (15,8))
axes = fig.add_axes([0,0,1,1])

axes.plot(df['data'], df['temperatura'], color = 'g')
axes.set_title('Temperatura no momento', fontsize = 25, pad = 20)
axes.set_xlabel('Data', fontsize = 20)
axes.set_ylabel('Temperatura', fontsize = 20)
axes.legend(['Temperatura'], loc = 'lower right', fontsize = 15)
axes.grid(True)

axes.axhline(max(df['temperatura']), color = 'black', linestyle = '--')
axes.axhline(min(df['temperatura']), color = 'black', linestyle = '--')
```

```python
#adicionando anotações
fig = plt.figure(figsize = (15,8))
axes = fig.add_axes([0,0,1,1])

axes.plot(df['data'], df['temperatura'], color = 'g')
axes.set_title('Temperatura no momento', fontsize = 25, pad = 20)
axes.set_xlabel('Data', fontsize = 20)
axes.set_ylabel('Temperatura', fontsize = 20)
axes.legend(['Temperatura'], loc = 'lower right', fontsize = 15)
axes.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

axes.annotate("Máximo", xy = (x1,y1), fontsize = 20)

axes.axhline(max(df['temperatura']), color = 'black', linestyle = '--')
axes.axhline(min(df['temperatura']), color = 'black', linestyle = '--')
```

```python
#adicionando anotações
fig = plt.figure(figsize = (15,8))
axes = fig.add_axes([0,0,1,1])

axes.plot(df['data'], df['temperatura'], color = 'g')
axes.set_title('Temperatura no momento', fontsize = 25, pad = 20)
axes.set_xlabel('Data', fontsize = 20)
axes.set_ylabel('Temperatura', fontsize = 20)
axes.legend(['Temperatura'], loc = 'lower right', fontsize = 15)
axes.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) - 5

axes.annotate("Máximo", xy = (x1,y1), fontsize = 20,
             xytext = (x2,y2), arrowprops = dict(facecolor = 'black'))

x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmin() - 7000]
y2 = min(df['temperatura']) + 5

axes.annotate("Mínimo", xy = (x1,y1), fontsize = 20,
             xytext = (x2,y2), arrowprops = dict(facecolor = 'black'))


axes.axhline(max(df['temperatura']), color = 'black', linestyle = '--')
axes.axhline(min(df['temperatura']), color = 'black', linestyle = '--')
```

- **Gráfico de barras**

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo','Segunda','Terça','Quarta','Quinta','Sexta','Sábado']
temperatura_por_dia_da_semana =  temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize = (5,4))
axes = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))

axes.bar(indice, temperatura_por_dia_da_semana)
axes.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
axes.set_xlabel('Dia da semana', fontsize = 15)
axes.set_ylabel('Temperatura média', fontsize = 15)
```

```python
fig = plt.figure(figsize = (5,4))
axes = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))

axes.bar(indice, temperatura_por_dia_da_semana)
axes.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
axes.set_xlabel('Dia da semana', fontsize = 15)
axes.set_ylabel('Temperatura média', fontsize = 15)

axes.set_xticks(indice)
axes.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize = (5,4))
axes = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'red', 'green','blue','yellow','orange','magenta']

axes.bar(indice, temperatura_por_dia_da_semana, color = cores, edgecolor = 'black')
axes.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
axes.set_xlabel('Dia da semana', fontsize = 15)
axes.set_ylabel('Temperatura média', fontsize = 15)

axes.set_xticks(indice)
axes.set_xticklabels(nome_dias)
```

- **Gráfico de pizza**

```python
fig = plt.figure(figsize = (5,4))
axes = fig.add_axes([0,0,1,1])

axes.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index)
axes.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
```

```python
#adicionando porcentagem
fig = plt.figure(figsize = (5,4))
axes = fig.add_axes([0,0,1,1])

axes.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index,
        autopct = '%.1f%%')
axes.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
```

```python
#destacar algumas fatias
fig = plt.figure(figsize = (5,4))
axes = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

axes.pie(temperatura_por_dia_da_semana, labels = temperatura_por_dia_da_semana.index,
        autopct = '%.1f%%', explode = explodir, shadow = True)
axes.set_title('Temperatura por dia da semana', fontsize = 15, pad = 10)
```
