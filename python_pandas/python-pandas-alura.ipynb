# Aula 6 - Tratando os dados faltantes

## Relatório de Análise V

```python
import pandas as pd
```

```python
#checar favores nulos
dados.isnull()
```

```python
dados = pd.read_csv('database/aluguel_residencial.csv', sep = ';')
```

```python
dados.notnull()
```

```python
dados.info()
```

```python
#chegando os valores nulos de uma determinada coluna
dados[dados['Valor'].isnull()]
```

```python
#elinar as linhas com valores nulos da coluna valores
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

Antes de eliminar as linhas com valores nulos em uma determinada coluna, é importante chegar que aquele dado faltanto é realmente necessário levando em consideração as outras informações da linha. 
Por exemplo, variavel condominio com valor nulo para uma linha referente a informações de casa não tem importancia, porque casa não tem condominio. Mas se for em relação a apartamento, pode ter um impacto, levando em consideração que a maioria dos apts tem condominio. 

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
#eliminou da base os aps com condominio nulo
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
#para os casos em que não precisa da info condominio, vamos substituir por zero.
dados.fillna(0,inplace=True)
```

```python
#ou
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

```python
dados.info()
```

```python
dados.to_csv('database/aluguel_residencial.csv', sep = ';', index = False)
```

## Identificando e removendo outliers

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))
```

```python
dados = pd.read_csv('database/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados.boxplot(['Valor'])
```

Mal da pra ver o boxplot por conta da interferência dos dados discrepantes.

```python
dados[dados['Valor'] >= 500000]
```

```python
valor = dados['Valor']
```

```python
#calculando as estatisticas do boxplot
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
#uma seleção dos dados que estão apenas dentro desses dois limites
#Tira os outliers da base
selecao = (valor >= limite_inferior) & (valor <= limite_superior)
```

```python
dados_new = dados[selecao]
```

```python
dados_new.boxplot(['Valor'])
```

Foi gerado um novo boxplot considerando somente os dados dentro dos limites. Ao gerar um novo ele recalcula as estatisticas novamente de forma automatica e acaba apontando novos outliers dentro daquela nova base. Não necessária esses novos valores fora dos limites estabelecidos são outliers, por isso é necessário conhecer sua base e o contexto dos valores. Aqui é posssível observar que ele aponto como outliers casas com alguel entre 10 e 14 mil, entretanto isso é possível e real. 
Uma sujestão é tratar esses dados de forma separada, separando os alugueis por classes, alto, médio e baixo padrão.

```python
dados.hist(['Valor'])
dados_new.hist(['Valor'])
```

```python
#Preenche cada gráfico de acordo com seu tipo

#gráfico 1 - scatterplot (dispersão)
g1.scatter(dados['Valor'], dados['Area'])
g1.set_title('Valor vs Area')

#gráfico 2 - histograma
g2.hist(dados['Valor'])
g2.set_title('Histograma')

#gráfico 3 - gráfico de linha, apresenta algumas amostras dos dados Valor
#crigando amostra de tamanho 100 aleatório
dados_g3 = dados['Valor'].sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

#gráfico 4 - barras
#agrupa os dados de valores por tipo de imóvel
grupo = dados.groupby('Tipo')['Valor']
#calcula a média de valores por tipo e define os paramentos do gráfico
labels = grupo.mean().index
valores = grupo.mean().values
g4.bar(labels, valores)
g4.set_title('Valor médio por tipo')
```
